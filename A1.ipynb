```python
# import pandas and numpy
import pandas as pd
import numpy as np
```

```python
# importing california drought dataset
cal_drought_data = pd.read_csv('drought dataset/full_dataset_CA_sorted.csv')

# dropping the unnecessary column
cal_drought_data = cal_drought_data.drop(columns=['Unnamed: 0'])

# viewing first 5 and last 5 rows of dataset
print(cal_drought_data.head())
print(cal_drought_data.tail())
```

```python
# importing soil data
soil_data = pd.read_csv('drought dataset/soil_data.csv')
print(soil_data.head())
```

```python
# viewing column names of the dataset
print(cal_drought_data.columns)
# viewing shape of the dataset
print(cal_drought_data.shape)
```

```python
# viewing summary statistics of the dataset
print(cal_drought_data.info())
#print(cal_drought_data.dtypes)
```

```python
print(cal_drought_data.describe())
```

```python
t1 = cal_drought_data[cal_drought_data.score.notnull()]
print(t1)
t2 = t1[cal_drought_data['fips']==6001]
print(t2)
t3 = cal_drought_data[cal_drought_data['fips']==6001]
print(t3)
```

```python
# preparing the weekly aggregated data
index_list = list(t2.index)
print(index_list)
y = t2.score
print(y)
x = t3.drop(columns=['fips','date'])
print(x)
print(x[0:4])
```

```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# plotting histograms of all the columns to see value distribution
cal_drought_data.hist(figsize=(20,20))
plt.show()
```

```python
# plotting box and whiskers plot to view outliers
cal_drought_data.plot(kind='box',figsize=(200,10),subplots=True,sharex=False,sharey=False)
plt.show()
```

```python
col_names = list(cal_drought_data.drop(columns=['fips','date','score']).columns)
print(col_names)
```

```python
fig, ax = plt.subplots(18,figsize=(30,100))
c = 0
for col in col_names:
    ax[c].scatter(x = cal_drought_data[col], y = cal_drought_data['score'])
    ax[c].set_xlabel(col)
    ax[c].set_ylabel("score")
    c = c+1
plt.show()
```
