```python
import pandas as pd
```

```python
# laod fips code and county data
fips = pd.read_excel('drought dataset/fips.xlsx')
```

```python
# view first 5 rows of the dataset
fips.head()
```

```python
# groupby state and count the number of counties in each state
fips.groupby('State').agg('count')
```

```python
# save a .csv file of grouped fips dataset
#fips.to_csv('drought dataset/fips.csv')
```

```python
# load validation dataset
df_validation = pd.read_csv('drought dataset/validation_timeseries/validation_timeseries.csv')
```

```python
# view first 5 rows of validation dataset
df_validation.head()
```

```python
# load test dataset
df_test = pd.read_csv('drought dataset/test_timeseries/test_timeseries.csv')
```

```python
# view first 5 rows of dataset
df_test.head()
```

```python
# load train dataset
df_train = pd.read_csv('drought dataset/train_timeseries/train_timeseries.csv')
```

```python
# view first 5 rows of train dataset
df_train.head()
```

```python
# merge train, validation and test
frames = [df_train,df_validation,df_test]
```

```python
full_dataset = pd.concat(frames)
```

```python
# viewing summary statistics of merged dataset or full dataset
full_dataset.info()
```

```python
# sorting the dataset according to fips codes
full_dataset_sorted = full_dataset.sort_values(by=['fips']).reset_index(drop=True)
# verifying the sorted dataset, 7671 records for each county should be present
print(full_dataset_sorted.head(7672))
print(len(full_dataset_sorted[full_dataset_sorted['fips']==1001]))
```

```python
# exporting the full dataset to csv
full_dataset_sorted.to_csv('drought dataset/full_dataset_sorted.csv')
```

```python
# retrieving fips code data for California or 'CA' counties
fips_CA = fips[fips['State']=='CA']
print(fips_CA)
# viewing summary statistics
print('\n',fips_CA.info())
```

```python
# retrieving CA counties data from full dataset
full_dataset_CA = full_dataset.loc[full_dataset['fips'].isin(fips_CA['FIPS'])]
# viewing the first 5 rows of CA counties dataset
print(full_dataset_CA.head())
# viewing summary statistics of CA county dataset
print(full_dataset_CA.info())
```

```python
full_dataset_CA.describe()
```

```python
# sorting CA counties dataset
full_dataset_CA_sorted = full_dataset_CA.sort_values(by=['fips']).reset_index(drop=True)
# verifying the sorting, each county should have 7671 rows of data
print(full_dataset_CA_sorted.head(7672))
print('\n',len(full_dataset_CA_sorted[full_dataset_CA_sorted['fips']==6001]))
```

```python
# exporting the dataset to a .csv file
full_dataset_CA_sorted.to_csv('drought dataset/full_dataset_CA_sorted.csv')
```

```python
# plotting bar graph of each column to view data distribution
import matplotlib.pyplot as plt
# getting the column name list for iteration
ind_list = list(full_dataset_CA.columns)
ind_list.remove('fips')
ind_list.remove('date')
print(ind_list)
#full_dataset_CA['PRECTOT'].hist(bins=5)
```

```python
for col_name in ind_list:
    plt.figure()
    plt.hist(full_dataset_CA[col_name])
    plt.xlabel(col_name)
```
